## Membaca Dataset csv dari Url

```python
# mengimpor paket pandas kemudian diberi nama alias pd
import pandas as pd
# mengimpor paket seaborn kemudian diberi nama alias sns
import seaborn as sns
# mengimpor paket numpy kemudian diberi nama alias np
import numpy as np
# mengimpor paket statistics kemudian diberi nama alias st
import statistics as st
```

```python
# membaca dataset csv dari url
url = 'https://raw.githubusercontent.com/ThreeYas/BMI/main/500_Person_Gender_Height_Weight_Index.csv'
data = pd.read_csv(url)
```

# Statistik Matematika

## Modus

```python
# Modus dari kolom Height
st.mode(data.Height)
```

```python
# Modus dari kolom Index
st.mode(data.Index)
```

## Median

```python
# Median dari kolom Height
st.median(data.Height)
```

```python
# Median dari kolom Weight
st.median(data.Weight)
```

```python
# Median dari kolom Index
st.median(data.Index)
```

## Mean

```python
# Rata - rata dari kolom Height
st.mean(data.Height)
```

```python
# Rata - rata dari kolom Weight
st.mean(data.Weight)
```

```python
# Rata - rata dari kolom Index
st.mean(data.Index)
```

## Kuartil

```python
# Kuartil dari kolom Height
np.percentile(data.Height, [25, 50, 75])
```

```python
# Kuartil dari kolom Weight
np.percentile(data.Weight, [25, 50, 75])
```

```python
# Kuartil dari kolom Index
np.percentile(data.Index, [25, 50, 75])
```

## Desil

```python
#Desil 1 - 10 dari kolom Height
np.percentile(data.Height, np.arange(0, 100, 10))
```

```python
#Desil 1 - 10 dari kolom Weight
np.percentile(data.Weight, np.arange(0, 100, 10))
```

```python
#Desil 1 - 10 dari dari Kolom Index
np.percentile(data.Index, np.arange(0, 100, 10))
```

## Range

```python
# Jangkaun dari kolom Height
# rumus jangkauan R = Xmax - Xmin
Xmax = max(data.Height)
Xmin = min(data.Height)
R = Xmax - Xmin
R
```

```python
# Jangkaun dari kolom Weight
# rumus jangkauan R = Xmax - Xmin
Xmax = max(data.Weight)
Xmin = min(data.Weight)
R = Xmax - Xmin
R
```

```python
# Jangkaun dari kolom Index
# rumus jangkauan R = Xmax - Xmin
Xmax = max(data.Index)
Xmin = min(data.Index)
R = Xmax - Xmin
R
```
